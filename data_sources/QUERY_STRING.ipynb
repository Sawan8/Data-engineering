A query string is the portion of a URL that contains key-value pairs appended after `?`, used to pass parameters to the server to control filtering, pagination, sorting, and other behaviors without altering the resource path.

`protocol://domain/path?query_string`

for eg:
<br> `https://api.example.com/products?limit=10&offset=20`

# Types of Query Parameter

### FILTERING PARAMETERS

Filtering parameters restrict the dataset returned by the API based on specific conditions.


```python
import requests

url = "https://jsonplaceholder.typicode.com/posts"

params = {
    "userId": 1
}

response = requests.get(url, params=params)

data = response.json()

print(response.url)
print(len(data))   # should be filtered
print(data[:2])
```

### SORTING PARAMETERS

Sorting parameters control the order in which data is returned by the API.

`/products?sort=price&order=asc`

*params = {
<br>    "sort": "price",
<br>    "order": "asc"
<br> }*

### FIELD SELECTION
Field selection parameters allow clients to specify which fields/columns should be returned in the response.

`/products?fields=id,name,price`

*params = {
<br>    "fields": "id,name,price"
<br>}*

### AGGREGATION / TRANSFORMATION PARAMETERS

Aggregation parameters allow summarizing or transforming data (e.g., count, sum, group).


```python
/orders?group_by=category&sum=price
```

**It is not seperated by many APIs, mainly BI tools API use these parameters**

for eg: google analytics API

### Boolean Parameters

Boolean parameters are query parameters that accept true or false values to enable or disable specific behaviors in an API.

Use Case: Show only available products in Ecommerce website

```python
Use Case: Show only available products
```

`/products?in_stock=true`

### Search Parameters

Search parameters are query parameters used to find records that match a given keyword or text pattern within one or more fields of a dataset.

Basic Structure: `?q=keyword`


```python
import requests

url = "https://dummyjson.com/products/search"

params = {
    "q": "phone"
}

response = requests.get(url, params=params)

data = response.json()

print(response.url)
print(data["products"][:2])
```

### Pagination

Pagination is a technique used to divide large datasets into smaller, manageable chunks so that data can be retrieved efficiently and incrementally from an API.
