In data engineering, we connect Python to MySQL when we want to:

<br> read data from MySQL tables
<br> run SQL queries
<br> load data into pandas
<br> validate source data

### Connection string format

```mysql+pymysql://username:password@host:port/database_name```

**Connecting Python to MySQL using SQLAlchemy means creating an engine that allows Python to execute SQL queries and read MySQL data in a structured way**


````python
from sqlalchemy import create_engine, text

host = "127.0.0.1"
port = "3306"
user = "root"
password = "sawan"
database = "hr"

connection_string = f"mysql+pymysql://{user}:{password}@{host}:{port}/{database}"
engine = create_engine(connection_string) 

with engine.connect() as conn: # creating connection
    result = conn.execute(text("SELECT * from hr.employees limit 10")) # text function helps in formatting sql queries
    for row in result:
        print(row)
````

### fetchone

fetchone() returns one row at a time from the result.
<br>When we run it multiple time then it will return next row

````python
from sqlalchemy import create_engine, text

connection_string = f"mysql+pymysql://{user}:{password}@{host}:{port}/{database}"
engine = create_engine(connection_string) 

with engine.connect() as conn:
    result = conn.execute(text("SELECT * FROM hr.employees limit 5"))
    row = result.fetchone()
    print(row)
````

````python
with engine.connect() as conn:
    result = conn.execute(text("SELECT * FROM hr.employees limit 5"))

    row1 = result.fetchone()
    print("First row:", row1)

    row2 = result.fetchone()
    print("Second row:", row2)
````

### Fetchall

fetching all records in a time

````python
with engine.connect() as conn:
    result = conn.execute(text("SELECT * FROM hr.employees limit 10"))
    rows = result.fetchall()

    for row in rows:
        print(row)
````

### Difference between engine.connect() and engine.begin()

Use **engine.connect()** for read-only queries like SELECT, where we just want to fetch data.
<br> Use **engine.begin()** for write operations like INSERT, UPDATE, and DELETE, because it handles transaction commit/rollback safely and is the better production practice for data changes.

### loading data to DataFrame

````python
import pandas as pd
df = pd.read_sql("SELECT * FROM hr.employees", engine)
````

### Loading data in chunks

````python
query = "SELECT first_name, last_name FROM hr.employees limit 4;"

chunks = []

for chunk_df in pd.read_sql(query, engine, chunksize=2): # dividing data into chunks
    print("This is one chunk:")
    print(chunk_df)
    print(type(chunk_df))
    chunks.append(chunk_df)

final_df = pd.concat(chunks, ignore_index=True)

print("Final combined dataframe:")
print(final_df)
````
