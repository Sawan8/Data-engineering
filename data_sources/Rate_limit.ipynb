### Rate limiting

Rate limiting means an API allows only a certain number of requests within a time window. This protects the service from overload, abuse, and unfair usage, and keeps the API available for everyone.


Pagination answers: `“How do I fetch all pages?”`
Rate limiting answers: `“How fast am I allowed to fetch them?”`

### Why APIs use rate limiting

APIs apply limits so one client does not consume too many resources. In real systems, this matters because thousands of users or services may be hitting the same backend at once.

### HTTP Status Code (Standard)

`429 Too Many Requests`

### Retry-After

Retry-After is an HTTP response header that tells the client how long to wait before making the next request.

It is usually sent when:

hit a rate limit (429 Too Many Requests)
The server is temporarily unavailable (503 Service Unavailable)

It can have two formats:
<br>**Seconds:**
<br> Retry-After: 120 → wait 120 seconds
<br>**Timestamp:**
<br> Retry-After: Wed, 21 Oct 2026 07:28:00 GMT

### Throttling (Definition)

Throttling is a proactive technique used to control the rate of outgoing requests from a client by limiting how frequently requests are sent to an API or system.

**Fixed delay throttling**

<br> Wait constant time between requests
<br> Example: 1 request every 2 seconds

**Rate-based throttling**

<br> Define requests per unit time
<br> Example: 60 requests per minute

**Adaptive throttling (advanced)**

<br> Adjust rate dynamically based on:

response time
<br> rate limit headers

```python
import requests
import time

url = "https://api.github.com/repos/python/cpython/issues"

page = 1

while page <= 3:
    params = {
        "page": page,
        "per_page": 5
    }

    response = requests.get(url, params=params)

    print(f"Page {page} | Status: {response.status_code}")

    data = response.json()

    for issue in data:
        print("Issue ID:", issue["id"])

    page += 1

    time.sleep(2)   # throttling between pages
```

### Backoff
Backoff is a reactive retry strategy in which a client intentionally delays subsequent retry attempts after a failed request, typically increasing the delay over time, to reduce load on the server and improve the likelihood of a successful response.

```python
url = "https://api.github.com/repos/python/cpython"

max_retries = 5

for attempt in range(max_retries):
    response = requests.get(url)

    if response.status_code == 200:
        print("Success:", response.json()["full_name"])
        break
    else:
        wait = 2 ** attempt   # exponential backoff
        print(f"Request failed. Retrying in {wait} seconds...")
        time.sleep(wait)
```

**Note**
Backoff means don’t retry immediately after failure
<br> wait, and increase the wait time if failures continue.
