## Parsing

Parsing a response is the process of reading, validating, and transforming raw API or system output (such as JSON, XML, or text) into structured, usable data formats (like dictionaries, lists, or DataFrames) for further processing, storage, or analysis.

### Never parse without checking response                                                                          

```python
import requests

url = "https://jsonplaceholder.typicode.com/users"

response = requests.get(url)
response.raise_for_status() # It will raise the error if got any unsuccessful code

if "application/json" not in response.headers.get("Content-Type", ""): # check if response is json
    raise ValueError("Not a JSON response")

data = response.json()
print(type(data))

```

### Always use .get() for safety

```python
user = data[0]

print(user.get("name"))
print(user.get("email"))
print(user.get("address", {}).get("city"))
```

**user.get("address", {})**
<br>
Tries to get "address"
<br> `If missing → returns {} (empty dict)`

```python
print(user.get("address1").get("city")) #Crashes if "address" is missing
```

### Loop Through API Response

```python
records = []

for user in data:
    records.append({
        "id": user.get("id"),
        "name": user.get("name"),
        "email": user.get("email"),
        "city": user.get("address", {}).get("city")
    })

print(records[:5])
```

```python
# Storing into dataframe

import pandas as pd

df = pd.DataFrame(records)
print(df.head())
```
