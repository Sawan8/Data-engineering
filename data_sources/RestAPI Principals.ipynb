REST (Representational State Transfer) is a set of architectural rules that define how a well-designed API should behave. Any API that follows these rules is called a RESTful API.

## LAYERED SYSTEM

client doesn't need to know whether it is talking directly to the actual server or to a cache server, or security gateway in between. The layers are invisible. This allows APIs to scale by adding infrastructure layers without changing the client.

## REPRESENTATION OF RESOURCES

In REST, resources are represented in a standard format (such as JSON or XML) that can be transmitted between client and server

## UNIFORM INTERFACE / CRUD

Uniform interface makes the API consistent and learnable. Every REST API uses the same standard methods (GET, POST, PUT, PATCH, DELETE), the same URL patterns (/resources/{id}), and the same format (JSON). Once you know one REST API, you roughly know all of them.

## CLIENT-SERVER ARCHITECTURE

It means the client (your app or pipeline) and the server (the API) are completely independent. The client only cares about getting data; the server only cares about storing and serving it. They communicate only through the API contract, neither knows how the other is built internally.

## STATELESSNESS

It is the most important principle. Every single request must contain all the information the server needs to process it: authentication token, parameters, everything. The server keeps zero memory between requests. There is no concept of "you logged in 5 minutes ago, so I remember you." This is why you send your Authorization header with every request, not just the first one.

## Cacheable
Cacheable means responses should declare whether they can be stored and reused. If a response is marked cacheable, the client can reuse it instead of hitting the server again. This reduces load and speeds things up dramatically for frequently-fetched data that doesn't change often.
