## What is an async call?

An async call means Python sends a request and, instead of sitting idle, it can start handling another request while waiting for the first response.

````python
import asyncio
import httpx

urls = [
    "https://jsonplaceholder.typicode.com/posts/1",
    "https://jsonplaceholder.typicode.com/posts/2",
    "https://jsonplaceholder.typicode.com/posts/3"
]

async def fetch(url, client):
    response = await client.get(url)
    response.raise_for_status()
    return response.json()

async def main():
    headers = {"x-api-key": "YOUR_API_KEY"} # defining authentication
    async with httpx.AsyncClient(headers=headers,timeout=10.0) as client:
        tasks = [fetch(url, client) for url in urls]
        results = await asyncio.gather(*tasks)

        for item in results:
            print(item)

await main()
````

``` async def fetch(): ``` It means this function will work in async style.

```response = await client.get(url)``` “Wait for this API response, but let other async tasks run meanwhile.”

```tasks = [fetch(url, client) for url in urls]```
<br> This is a list of tasks (not executed yet)

```results = await asyncio.gather(task1, task2, task3)```
<br> Takes multiple async tasks
<br> Runs them concurrently
<br> Waits until all finish
<br> Returns results in same order

##### What does *tasks mean?
*tasks unpacks a list of tasks into separate arguments so that asyncio.gather() can execute them concurrently.


### Limit the number of requests

````python


user_ids = [1, 2, 3, 4, 5]
semaphore = asyncio.Semaphore(2) # limit the request to 2

async def fetch_user(user_id, client):
    url = f"https://jsonplaceholder.typicode.com/users/{user_id}"

    async with semaphore:
        response = await client.get(url)
        response.raise_for_status()
        return response.json()

async def main():
    headers = {"x-api-key": "YOUR_API_KEY"} # defining authentication
    async with httpx.AsyncClient(headers=headers,timeout=10.0) as client:
        tasks = [fetch_user(user_id, client) for user_id in user_ids]
        users = await asyncio.gather(*tasks)

        for user in users:
            print(user["id"], user["name"])

await main()
````
