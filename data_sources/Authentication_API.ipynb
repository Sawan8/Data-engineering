### API Key Authorization

<br> You are given a unique key
<br> You send it with every request
<br> Server checks if key is valid

````python
import requests

url = "https://api.openweathermap.org/data/2.5/weather"

params = {
    "q": "Haldwani",
    "appid": "6627e7acc92527c3ba18bb0d8d6bf42d"
}

response = requests.get(url, params=params)

print(response.status_code)
print(response.text) 
````

**Step 1** : go to https://openweathermap.org
<br> **Step2** : create account and verify it through email verification
<br> **step3** : wait for few hours to get key activated


<span style="color:red"> **Exposing API key is strictly not allowed in industry**</span>

### Bearer Token Authorization

Bearer Token Authorization is a method where a client sends an access token in the request header to authenticate and authorize access to protected API resources.

<br> **Important Points:**
<br> Token is always shared in header
<br> Token has always expiry date

```python
url = "https://api.example.com/data"

headers = {
    "Authorization": "Bearer your_token"
}

response = requests.get(url, headers=headers)

print(response.status_code)
print(response.json())
```

### Basic Authentication
Basic Authentication is a method where a client sends a username and password with each request, encoded in Base64, to authenticate with an API.

```python
import requests

url = "https://api.example.com/data"

response = requests.get(url, auth=(username, password) )

print(response.status_code)
print(response.text)
```

### OAuth 2.0

OAuth 2.0 lets you access someone’s data using a token instead of their password.
##### Usually we get token through:

Step-by-step:
