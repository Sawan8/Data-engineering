## What is a Data Source?

A data source is any place where data is stored or generated that you need to collect.

When you order on **Swiggy**      →  your order goes into a DATABASE
<br> When you pay via **UPI**          →  payment app sends data via an API
<br> When HR sends a **salary sheet**  →  that is an EXCEL FILE
<br> When you browse **Instagram**     →  your clicks are saved as LOG FILES

## The Three Types of Data


### Type 1 — Structured Data

Data that is organised in rows and columns with a fixed, predefined structure. Every record looks exactly the same.

<br> Bank transactions table    →  every row has: account, amount, date
<br> Hospital patient records   →  every row has: patient_id, name, age
<br> School exam results        →  every row has: roll_no, subject, marks
<br> Employee salary sheet      →  every row has: emp_id, name, salary

**Where it lives:**
<br> Databases  →  PostgreSQL, MySQL (like Excel but for millions of rows)
<br> Files      →  CSV files, Excel sheets

### Type 2 —  Semi-Structured Data
Data that has some structure but is not organised in fixed rows and columns. It can be nested and fields can vary between records.

**JSON is most common Semi-strcutured data**

**Example**:
<br> {
<br>  "name": "Sawan",
<br>  "age": 28,
<br>  "skills": ["Python", "SQL", "AI"]
<br>}

### Type 3- Unstructured Data

Unstructured data is data that has no predefined structure, no fixed rows, no columns, and no consistent format. It is raw content in its natural form that cannot be stored directly in a traditional table or queried with SQL.

| TYPE          | REAL EXAMPLE                     |
|---------------|----------------------------------|
| PDF documents | Aadhaar card, bank statement     |
| Images        | Product photos, ID card scan     |
| Audio         | Call centre recordings           |
| Video         | CCTV footage, YouTube video      |

## Types Of Data Sources

### Databases
A database source is a system where data is stored in a structured or organized format and from which data can be accessed, queried, and extracted for use in applications or data pipelines.
<br> Examples → PostgreSQL, MySQL, Oracle, SQL Server

### APIs 

An API (Application Programming Interface) is a set of rules and protocols that allows different software applications to communicate with each other and exchange data.

### Files
In a file-based system, data is saved in files (like CSV, Excel, text), and programs read or write these files directly instead of using a database.
