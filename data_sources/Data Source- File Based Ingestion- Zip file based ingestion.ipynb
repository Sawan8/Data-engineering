### ZIP File with Multiple Files

```python
import zipfile

zip_path = "E:/data/Documents/Data_Engineering/Module1_Data_Source/module1-dataset/daily_files_multiple_csv.zip"

with zipfile.ZipFile(zip_path, "r") as zip_ref:
    print(zip_ref.namelist())
```

#### Reading specific file

```python


with zipfile.ZipFile(zip_path, "r") as zip_ref:
    with zip_ref.open("orders.csv") as file:
        df = pd.read_csv(file)

print(df.head())
```

#### Reading all files



```python
dataframes = {}

with zipfile.ZipFile(zip_path, "r") as zip_ref:
    for file_name in zip_ref.namelist():
        if file_name.endswith(".csv"):
            with zip_ref.open(file_name) as file:
                dataframe_name = file_name.removesuffix(".csv")
                print(dataframe_name)
                dataframes[dataframe_name] = pd.read_csv(file)


orders_df = dataframes["orders"]
customers_df = dataframes["customers"]
products_df = dataframes["products"]
```
