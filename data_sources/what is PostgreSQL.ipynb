## Comparison Table

| Category              | MySQL                              | PostgreSQL                                      |
|----------------------|------------------------------------|-------------------------------------------------|
| Type                 | Relational DBMS                    | Object-Relational DBMS (More advance data types)                        |
| ACID Compliance      | Yes (with InnoDB)                  | Yes                 |
| SQL Compliance       | Partial                            | Very high                                       |
| JSON Support         | Basic JSON type                    | Superior — JSON + JSONB (indexed)               |
| Array Support        | No                                 | Yes (native arrays)                             |
| Custom Data Types    | Limited                            | Extensive                                       |
| Full-text Search     | Basic                              | Advanced                                        |
| Indexing             | B-Tree, Hash, Full-text            | B-Tree, Hash, GiST, GIN, BRIN, SP-GiST          |
| Concurrency Model    | Row-level locking                  | MVCC — readers never block writers              |
| Ease of Setup        | Simpler, beginner-friendly         | More configuration needed                       |
| Best For             | Web apps, CMS, high-read workloads | Analytics, ETL, complex data workloads          |

##  Key Concepts Cheat Sheet

### ACID — The 4 Promises of a Reliable Database
```
A = Atomicity   → All or nothing. No half-saves.
C = Consistency → Data is always valid. No broken rules.
I = Isolation   → Parallel transactions don't interfere. Multiple transactions should not mess with each other
D = Durability  → Once saved, always saved. Even after a crash.
```

### MVCC — Multi-Version Concurrency Control
```
Problem:  1000 users reading/writing at the same time → chaos
Solution: Each transaction sees its own "snapshot" of the data
Result:   Readers never block writers. Writers never block readers.
Analogy:  Google Docs — multiple editors, no file locking
```

### JSON vs JSONB
```
JSON  → Stored as plain text. Slower to search.
JSONB → Stored in binary format. Pre-processed. Indexed.
        Searching inside JSONB = much faster.

```
