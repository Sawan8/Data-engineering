## 2. Create SparkSession

`appName()` gives a name to our Spark application.

`getOrCreate()` creates a new SparkSession or reuses an existing one.

````python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("pyspark") \
    .master("local[1]") \
    .config("spark.python.worker.reuse", "false") \
    .config("spark.driver.host", "127.0.0.1") \                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                    
    .config("spark.driver.bindAddress", "127.0.0.1") \
    .getOrCreate()
````

| Code | Explanation |
|---|---|
| `.master("local[1]") \` | Runs Spark on the local computer using 1 worker thread. This is safe for Windows setup. |
| `.config("spark.python.worker.reuse", "false") \` | Tells Spark not to reuse the same Python worker process. This helps avoid Python worker timeout issues on Windows. |
| `.config("spark.driver.host", "127.0.0.1") \` | Sets the Spark driver host to localhost. `127.0.0.1` means the current computer. |
| `.config("spark.driver.bindAddress", "127.0.0.1") \` | Tells Spark driver to listen on localhost only. This avoids hostname or network adapter issues. |
| `.getOrCreate()` | Creates a new SparkSession if it does not exist, or reuses the existing one if already available. |

## 5. Create Sample Data

We will create a small orders dataset manually.

Each row has:

```text
order_id, customer_name, city, amount
```

````python
    data = [
        (1, "Amit", "Delhi", 500),
        (2, "Priya", "Mumbai", 700),
        (3, "Rahul", "Delhi", 300),
        (4, "Sneha", "Pune", 900),
        (5, "Neha", "Mumbai", 400)
    ]
    
    columns = ["order_id", "customer_name", "city", "amount"]
````

## 6. Create First DataFrame

A DataFrame is like a table with rows and columns.

````python
df = spark.createDataFrame(data, columns)
````

## 7. Display Data using `show()`

````python
df.show()
````

## 8. Display Limited Rows

`show(3)` displays only 3 rows.

````python
df.show(3)
````

## 9. Show Full Values Without Truncation

`truncate=False` prevents long values from being shortened.

````python
df.show(truncate=False)
````

## 10. Check Schema using `printSchema()`

Schema means the structure of the DataFrame.

It shows:


````python
df.printSchema()
````

## 11. Check Column Names

## 12. Count Rows

`count()` returns the number of rows.

````python
df.count()
````

## Important Note

| Command | Meaning |
|---|---|
| `show()` | Displays rows |
| `show(3)` | Displays first 3 rows |
| `printSchema()` | Shows structure |
| `columns` | Shows column names |
| `count()` | Counts rows |

`show()` and `count()` are actions because they trigger Spark execution.
