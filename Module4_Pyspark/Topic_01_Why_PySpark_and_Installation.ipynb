# 1. What is PySpark?

**PySpark = Python + Apache Spark**

Apache Spark is a distributed data processing engine.  
PySpark allows us to use Spark using Python.

Simple definition:

> PySpark is used to process large data using Python, where the work can be divided across multiple machines.

In normal Python or Pandas, most work happens on one machine.  
In PySpark, the work can be distributed across many worker machines.

# 2. Why do we need PySpark?

Imagine this data engineering task:

```text
You have 500 GB sales data.
You need to clean it, join it with customer data,
calculate city-wise sales, and save the final output.
```

Using normal Python or Pandas, this may be very slow or may fail because the data may not fit in one machine's memory.

Spark solves this by dividing the data into smaller parts called **partitions**.

```text
500 GB data
   ↓
Divided into partitions
   ↓
Different workers process different partitions
   ↓
Final result is combined
```

This is called **distributed processing**.

# 3. Easy Real-Life Example

Suppose one teacher has to check **10,000 answer sheets**.

If one teacher checks all sheets alone, it will take a lot of time.

But if the work is divided among 20 teachers:

```text
Teacher 1 checks 500 sheets
Teacher 2 checks 500 sheets
Teacher 3 checks 500 sheets
...
Teacher 20 checks 500 sheets
```

The work finishes faster.

This is the same idea behind Spark.

Spark divides a large data processing job into smaller tasks and executes them in parallel.

# 4. Pandas vs PySpark

| Point | Pandas | PySpark |
|---|---|---|
| Data size | Small to medium | Large data |
| Processing | Single machine | Multiple machines |
| Memory | Local RAM | Distributed memory |
| Speed | Very good for small data | Better for big data |
| Main use | Analysis and small data work | Data engineering and big data pipelines |

Simple line:

> Pandas is like one person doing the work.  
> PySpark is like a team doing the work together.

# 5. SQL vs PySpark

SQL is excellent for querying structured data inside databases or data warehouses.

Example SQL query:

```sql
SELECT city, SUM(amount)
FROM sales
GROUP BY city;
```

But in data engineering, data may come from many places:

```text
CSV files
JSON files
Parquet files
Delta tables
APIs
Databases
Kafka streams
Cloud storage like S3, ADLS, or GCS
```

PySpark can read from many sources, transform the data, and write the final output.

# 7. Typical Data Engineering Flow with PySpark

A modern PySpark data pipeline often follows the **Bronze → Silver → Gold** pattern.

```text
Raw Data Sources
CSV / JSON / Database / API / Kafka
        ↓
Bronze Layer
Store raw data as-is
        ↓
Silver Layer
Clean data, remove duplicates, handle nulls
        ↓
Gold Layer
Business-ready tables for reporting and analytics
        ↓
Dashboard / ML / Business Users
```

PySpark is mainly used to transform raw data into clean and business-ready data.

# 8. Important Terms

### Environment Variable Setup for Java on Windows

To run PySpark locally, Java should be installed and the environment variables should be configured properly.

PySpark uses Python, but internally Apache Spark runs on the **Java Virtual Machine (JVM)**.  
That is why we need to set `JAVA_HOME`.

---

##### Step 1: Find Java Installation Path

Open **File Explorer** and go to:

```text
C:\Program Files\Java
```

Inside this folder, you may see a JDK folder like:

```text
jdk-17
jdk-21
```

Example Java installation path:

```text
C:\Program Files\Java\jdk-21
```

This path will be used as `JAVA_HOME`.

---

## Step 2: Open Environment Variables Window

Follow these steps:

```text
Windows Search
→ Type "Environment Variables"
→ Click "Edit the system environment variables"

→ Click "Environment Variables..."
```
<br> **always choose option where system environment variable is written** 
<br> You will see two sections:

```text
User variables
System variables
```

If you have admin access, use **System variables**.  
If you do not have admin access, use **User variables**.

---

## Step 3: Create `JAVA_HOME`

Under **System variables** or **User variables**, click:

```text
New
```

Enter:

```text
Variable name: JAVA_HOME
Variable value: C:\Program Files\Java\jdk-25.0.3
```

Use your actual JDK folder path.

For example:

```text
C:\Program Files\Java\jdk-17
```

or

```text
C:\Program Files\Java\jdk-21
```

or

```text
C:\Program Files\Java\jdk-25
```

Then click:

```text
OK
```

---

## Important Rule

`JAVA_HOME` should point to the **main JDK folder**, not the `bin` folder.

Correct:

```text
C:\Program Files\Java\jdk-25.0.3
```

Wrong:

```text
C:\Program Files\Java\jdk-25\bin
```

---

## Step 4: Add Java to `Path`

In the same **Environment Variables** window:

```text
Select Path
→ Click Edit
→ Click New
```

Add this value:

```text
%JAVA_HOME%\bin
```

Then click:

```text
OK
→ OK
→ OK
```

---
