### What is origin/main?

There are three related names:

main
<br> origin
<br> origin/main

**main** main = branch on our laptop
<br> **origin** This is the nickname of our GitHub repository.
<br> **origin/main**: Git's local memory of (online portal) GitHub's main branch

### Fetch vs Pull

#### fetch

When we run git fetch then it will update origin/main but not the main (working branch)





``` bash
git fetch origin

#### Pull
Git updates origin/main and also updates our local main (working branch)

``` bash
git pull

#### removing origin

``` bash
git remote remove origin
