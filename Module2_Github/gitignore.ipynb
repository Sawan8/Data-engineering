### .gitignore


In Git, `.gitignore` is a special file where we write the names or patterns of files/folders that Git should not track.

In simple words:

*.gitignore tells Git: “Do not add these files into version control*

**For example, we don't want to track:**
<br> temporary files
<br> log files
<br> secret files
