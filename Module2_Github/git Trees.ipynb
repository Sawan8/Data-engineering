In Git, project is managed through three important areas, often called the three trees. Understanding these three areas is very important because almost every Git command works by moving changes from one area to another.

The three trees are:

<br> Working Directory
<br> Staging Area
<br> Repository

!

### Working Directory

The working directory is the folder where we actually create, edit, and delete files.

For example, when we open  project folder in VS Code and create a file like student.txt.

At this stage, Git can see that something has changed, but Git has not saved it permanently yet.

### Staging Area

The staging area is a temporary preparation area.

Before making a commit, we tell Git exactly which changes we want to include in the next commit. 

### Repository

The repository is where Git permanently stores committed versions of your project.
