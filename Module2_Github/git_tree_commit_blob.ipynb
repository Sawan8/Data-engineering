Git does not simply store your project as normal files and folders. Git stores your project as internal objects. These objects help Git remember file content, folder structure, and project history.

<br> The three most important Git objects are:
<br> **Blob**   → stores file content
<br> **Tree**   → stores folder and file structure
<br> **Commit** → stores project snapshot and history


## Git as a Content-Addressed System

Git is called a content-addressed system because Git identifies and stores data based on the content itself, not mainly by the file name or file location.

In a normal file system, we usually identify a file by its name and path.

Git looks at the content inside the file, calculates a unique value from that content, and uses that value as the identity of the object.

That unique value is called a **hash**

### What is a Hash?

A hash is like a digital fingerprint of some content.
Git calculates a unique hash for that content. If the content changes even slightly, the hash changes.

So if the content is:

``` Hello Git``` 

is different from 

``` Hello GIT !```
