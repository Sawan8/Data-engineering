## Merge in Git

+ A merge in Git is the process of combining changes from one branch into another branch.
+ In real projects, we usually create a separate branch to work on a new feature, bug fix, experiment, or task so that the main branch remains stable.
+ Once the work on that separate branch is complete, we bring those changes back into the main branch using merge.
+ For example, if we created a branch called feature-login from main, did some work there, and later want that work to become part of main. we use merge

### Fast Forward Merge


+ A fast-forward merge is the simplest type of merge in Git. It happens when the current branch has not received any new commit after the feature branch was created. 
+ In this situation, the feature branch is simply ahead of the current branch. Since there is no separate work on the current branch, Git does not need to create a new merge commit. It simply moves the current branch pointer forward to the latest commit of the feature branch. That movement of the branch pointer is called a fast-forward merge.
