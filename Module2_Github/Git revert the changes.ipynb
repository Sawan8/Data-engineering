### Situation 1: Change is only in Working Directory

You edited a file, but you have not staged it yet.

``` bash
git status 
```

*you will see file changes*

``` bash
git restore pipeline.py
```

``` git restore . ``` will restore all files in one go

### Situation 2: Change is already staged

``` bash
git restore --staged filename
```

``` bash
git restore filename
```

### Situation 3: Change is already committed

``` bash 
git revert commit_id
```

**VS code window will pop up after running this command, you can CTRL + S and close the editor.**

``` bash
git log --oneline

git status
```

**restore** = undo changes before commit

**revert** = undo a committed change safely
