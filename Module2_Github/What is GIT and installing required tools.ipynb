### What is Git?
<br> Git is a version control system. It is used to track changes in project files over time. In real projects, code is not written only once. We keep changing it, improving it, fixing bugs, adding new logic, and sometimes removing old code. Git helps us manage all these changes properly.

Git is a tool that maintains the **history of your project**

### Why Git is Needed?
<br> Without Git, we would have many versions of files


```
project/
│
├── extract.py
├── extract_final.py
├── extract_final_new.py
├── extract_latest.py
```

With Git, we keep one clean project folder, and Git manages the history internally.

### Commit

Git works like a checkpoint system. Whenever your project reaches a meaningful stage, you save that stage in Git. This saved checkpoint is called a commit.

```
Project Timeline

Commit 1
        ↓
Commit 2
        ↓
Commit 3
        ↓
Commit 4
```

### Git vs Github

Git is the tool installed on our system. It tracks project history locally. GitHub is an online platform where Git repositories are uploaded and shared with others.
