## Git- Branch

In Git, a branch is used when we want to work on a new feature, bug fix, experiment, or change without disturbing the main working code.

In real data engineering projects, we do not directly change the production code.

**Step1:** go to folder
<br>

**step2:** check on which branch you are working
``` bash

**step3**: creating new branch
``` bash
git branch feature/demo

**step4** use git branch to check what branches are available

**step5** Switch branch using git checkout

``` bash
git checkout feature/demo

**step6** go to vs code and create new file

**step7** commit the file

``` bash

git add .
git commit --m "dummy file commit"


**step8** now switch back to main branch, then you will notice "dummy.py" file is not available

``` bash
git checkout master

### Create and switch branch together using checkout -b

Instead of two commands:

``` bash
git branch feature/demo
git checkout feature/demo

**we can use**

``` bash
git checkout -b feature/demo

### deleting the branch

``` bash
git branch -d feature/demo

**Note** 
<br> 1. we cant delete the current branch, you have to switch to another branch
<br> 2. if you want to delete any branch forcefully then use "-D"
