#### Check Docker system information

``` bash
docker info
```

This command gives detailed information about Docker installation.

#### Running test container using test image

``` bash
docker run hello-world
```


#### check running container

``` bash
docker ps
```

This command shows only currently running containers.

#### Check all containers

``` bash
docker ps -a

```

This command shows all containers:

<br> Running containers
<br> Stopped containers
<br> Exited containers

#### Check downloaded images

``` bash
docker images
```
This command shows images available on your local system.

**Note: hello-world image immediately stopped**

### lets try with real image

##### lets run the nginx image

``` bash
docker run nginx
```

**Note: Docker is showing live logs in the same terminal, it is running in foreground**

<span style="color:red">To stop the foreground container use `ctrl + c`</span> 

#### run in background

```
docker run -d nginx
```

| Part         | Meaning                         |
| ------------ | ------------------------------- |
| `docker run` | Create and start container      |
| `-d`         | Detached mode / background mode |
| `nginx`      | Image name                      |



**Now Docker will return a container ID and our terminal will be free**



check running container:
``` bash
docker ps
```

| Command             | What it does                            | Creates container? | Starts application? |
| ------------------- | --------------------------------------- | -----------------: | ------------------: |
| `docker pull nginx` | Downloads image only                    |                 No |                  No |
| `docker run nginx`  | Creates and starts container from image |                Yes |                 Yes |
