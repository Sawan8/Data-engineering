# 1. What is Docker Compose?

Docker Compose is a tool used to manage Docker containers using a YAML file.

In simple words:

> Docker Compose allows us to run one or more Docker containers using a single configuration file.

That configuration file is usually named:

```text
compose.yml
```

Earlier, we used commands like:

```bash
docker run nginx
docker run -d -p 8080:80 nginx
docker ps
docker stop <container_id>
```

But when we have multiple containers, writing long commands again and again becomes difficult.

Docker Compose solves this problem.

# 2. Why Docker Compose is Needed?

Suppose we have an application with multiple parts:

```text
Nginx container
Python API container
Database container
Worker container
```

Without Docker Compose, we need to manually run many Docker commands.

Example:

```bash
docker run -d --name nginx -p 8080:80 nginx
docker run -d --name python-api -p 8000:8000 python-api-image
```

With Docker Compose, we define everything once in `compose.yml` and run:

```bash
docker compose up
```

To stop everything:

```bash
docker compose down
```

# 3. Docker CLI vs Docker Compose

| Point | Docker CLI | Docker Compose |
|---|---|---|
| Main use | Run/manage containers manually | Manage containers using YAML file |
| Best for | One container practice | Multi-container applications |
| Configuration | Written in command line | Written in `compose.yml` |
| Repeatability | Harder | Easier |
| Example | `docker run nginx` | `docker compose up` |

Simple understanding:

```text
Docker CLI = manual commands
Docker Compose = configuration-based container management
```

# 4. What is `compose.yml`?

`compose.yml` is the main file where we define containers/services.

A basic Compose file looks like this:

```yaml
services:
  nginx:
    image: nginx
    ports:
      - "8080:80"
```

| Line | Meaning |
|---|---|
| `services:` | Section where we define containers |
| `nginx:` | Name of the service |
| `image: nginx` | Use ready-made nginx image |
| `ports:` | Expose container port to laptop |
| `"8080:80"` | Laptop port 8080 maps to container port 80 |

# 5. Important Docker Compose Commands

| Command | Meaning |
|---|---|
| `docker compose up` | Start services and show logs |
| `docker compose up -d` | Start services in background |
| `docker compose up --build` | Build images and start services |
| `docker compose ps` | Show Compose services |
| `docker compose logs` | Show service logs |
| `docker compose down` | Stop and remove containers/network |

Most commonly used commands:

```bash
docker compose up --build
docker compose up -d
docker compose ps
docker compose logs
docker compose down
```

# 8. Project Structure for Multiple Containers

Create this folder structure:

Meaning:

| File/Folder | Purpose |
|---|---|
| `compose.yml` | Main Compose configuration |
| `python-api/` | Folder for Python API service |
| `python-api/Dockerfile` | Instructions to build Python API image |
| `python-api/requirements.txt` | Python packages required |
| `python-api/main.py` | FastAPI application code |

# 9. Create Project Folder

Run these commands:

```bash
mkdir compose-multiple-demo
cd compose-multiple-demo
mkdir python-api
```

# 10. Create Python API File

Create this file:

```text
python-api/main.py
```

Write this code:

```python
from fastapi import FastAPI

app = FastAPI()

@app.get("/")
def home():
    return {"message": "Hello from Python API"}
```

When we open:

```text
http://localhost:8000
```

we should get:

```json
{"message":"Hello from Python API"}
```

# 11. Create `requirements.txt`

Create this file:

```text
python-api/requirements.txt
```

Write:

```text
fastapi==0.115.6
uvicorn==0.34.0
```

`fastapi` is used to create the API.  
`uvicorn` is used to run the API server.

# 12. Create Dockerfile for Python API

Create this file:

```text
python-api/Dockerfile
```

Write:

```dockerfile
FROM python:3.11-slim

WORKDIR /app

COPY requirements.txt .

RUN pip install -r requirements.txt

COPY main.py .

CMD ["uvicorn", "main:app", "--host", "0.0.0.0", "--port", "8000"]
```

# 13. Dockerfile Explanation

| Dockerfile Line | Meaning |
|---|---|
| `FROM python:3.11-slim` | Use Python 3.11 lightweight image as base |
| `WORKDIR /app` | Create/use `/app` folder inside container |
| `COPY requirements.txt .` | Copy requirements file into container |
| `RUN pip install -r requirements.txt` | Install Python packages from requirements file |
| `COPY main.py .` | Copy API code into container |
| `CMD [...]` | Start FastAPI app when container starts |

Important:

```text
RUN = executes during image build
CMD = executes when container starts
```

# 14. Create Final `compose.yml`

Create this file in the main folder:

```text
compose.yml
```

Write:

```yaml
services:
  nginx:
    image: nginx
    container_name: demo-nginx
    ports:
      - "8080:80"

  python-api:
    build:
      context: ./python-api
      dockerfile: Dockerfile
    container_name: demo-python-api
    ports:
      - "8000:8000"
```

# 15. `compose.yml` Explanation

## `services`

```yaml
services:
```

This is where we define all containers.

In this example, we define two services:

```text
nginx
python-api
```

Each service creates one container.

---

## `image`

```yaml
image: nginx
```

This means:

> Use ready-made Nginx image from Docker Hub.

Use `image` when you do not need to create your own image.

---

## `build`

```yaml
build:
  context: ./python-api
  dockerfile: Dockerfile
```

This means:

> Build custom image using Dockerfile inside `python-api` folder.

| Key | Meaning |
|---|---|
| `context: ./python-api` | Folder used for build |
| `dockerfile: Dockerfile` | Dockerfile name inside that folder |

Use `build` when you have your own code and Dockerfile.

# 16. `container_name` and `ports`

## `container_name`

```yaml
container_name: demo-python-api
```

This gives a readable name to the container.

## `ports`

```yaml
ports:
  - "8000:8000"
```

This means:

```text
Laptop port 8000 → Container port 8000
```

So we can open the API using:

```text
http://localhost:8000
```

For Nginx:

```yaml
ports:
  - "8080:80"
```

This means:

```text
Laptop port 8080 → Nginx container port 80
```
