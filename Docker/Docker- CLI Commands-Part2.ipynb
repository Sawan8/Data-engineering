#### Run container with a name

``` bash
docker run -d --name my-first-container nginx
```


#### Check whether your container starts with this name

``` bash
docker ps
```

2. Stop and remove named container

``` bash
docker stop my-first-container
```

``` bash
docker rm my-first-container
```

#### Run Nginx with port mapping

``` bash
docker run -d --name my-first-container -p 8080:80 nginx
```

| Part                  | Meaning                                       |
| --------------------- | --------------------------------------------- |
| `-d`                  | Run container in background                   |
| `--name my-nginx-web` | Give container a name                         |
| `-p 8080:80`          | Map laptop port `8080` to container port `80` |
| `nginx`               | Image name                                    |


**Now open browser**
``` bash
http://localhost:8080
```

**We can observe that image is running**

#### 5. Check container logs

``` bash
docker logs my-first-container
```

*to continously follow logs*
```
docker logs -f my-first-container
```

#### 6. Go inside running container
When we say go inside a running container, it means we open a terminal/shell inside the container environment


``` bash
docker exec -it my-first-container bash
```

| Part           | Meaning                              |
| -------------- | ------------------------------------ |
| `docker exec`  | Run command inside running container |
| `-it`          | Interactive terminal mode            |
| `my-first-container` | Container name                       |
| `bash` / `sh`  | Shell inside container               |


**To come back outside the container, run:**
``` bash
 exit
 ```

#### we can use container id instead of container name
``` bash
docker exec -it ec1115303e84 bash
```
