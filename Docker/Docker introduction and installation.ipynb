### what is docker?
In software development and data engineering, one common problem is that code works on one machine but fails on another machine. This usually happens because every system has a different setup, such as a different Python version, different library versions, missing dependencies, or different operating system settings.

Docker solves this problem by creating a fixed and repeatable environment for the application. Once an application is packaged with Docker, it can run in the same way on a laptop, server, or cloud platform.

### What is container?

Docker is a platform used to package and run applications inside containers. A container includes the application code, required libraries, dependencies, and runtime environment.

### Image and container basic idea

An image is like a blueprint or template. It contains everything needed to run an application.

A container is the running version of that image. When we run an image, Docker creates a container from it.

### Docker vs Virtual Machine

A virtual machine runs a full operating system inside another system. Because of this, virtual machines are heavy and take more time to start.

Docker containers share the host operating system kernel and package only the application, required libraries, dependencies, and configuration. Because of this, containers start quickly, use fewer resources, and make deployment easier. For example, instead of creating a full Ubuntu VM and manually installing Python, PostgreSQL, or Airflow, we can run them directly as Docker containers

### Docker Installation

Before installation, check these:
<br> **STEP 1:** Open Task Manager → Performance → CPU → Virtualization: Enabled 
<br> **if disabled then go to bios settings and enable it**

Inside BIOS/UEFI, search for one of these options and enable it:

``` bash
Processor	Setting name
Intel	    Intel Virtualization Technology / VT-x
AMD	SVM    Mode / AMD-V 
```

Change it from:

Disabled → Enabled

Then save and exit, usually with:

F10 → Save & Exit
<br>
<br> **STEP 2:** Install WSL 2 on Windows

+ Open PowerShell as Administrator and run: (Reboot your system once done)
  ``` bash
  wsl --install
  
  wsl --version
  ```
<br> **Step 3:** download it from below URL:
*https://www.docker.com/products/docker-desktop*


**Download for Windows — AMD64**

<br> Even if your processor is Intel, we can choose AMD64. Here, AMD64 means 64-bit Windows for Intel/AMD processors, not only AMD brand.

Choose ARM64 only if your laptop has an ARM processor
<br> 

**IMP**
```bash
During installation, keep this option selected:

Use WSL 2 instead of Hyper-V
```
