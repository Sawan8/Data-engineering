If two containers are in the same Docker network, one container can talk to another container using its container name.

**Step 1: Create a Docker network**

``` bash
docker network create demo-network
```

**Step 2: Run Nginx container inside this network**
``` bash
docker run -d --name nginx-server --network demo-network nginx
```

| Part                     | Meaning                         |
| ------------------------ | ------------------------------- |
| `docker run -d`          | Run container in background     |
| `--name nginx-server`    | Give container name             |
| `--network demo-network` | Attach container to our network |
| `nginx`                  | Use Nginx image                 |


**Step 3: Run curl container in the same network**
``` bash
docker run --rm --network demo-network curlimages/curl http://nginx-server
```

**Step 4: Try without network**

Now let’s prove why network matters.

Run a curl container without using demo-network:

``` bash
docker run --rm curlimages/curl http://nginx-server
```
