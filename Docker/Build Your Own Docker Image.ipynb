**What is Dockerfile?**

A Dockerfile is a text file that contains step-by-step instructions to build a Docker image.

**Step 1: create folder**

``` bash
mkdir docker-python-demo
cd docker-python-demo
```

**step2: Create Python file using vs code**



**Step 3: Create Dockerfile**

Now create one more file in the same folder.

File name should be exactly: `Dockerfile`

+ adding step1 in Dockerfile, Use Python 3.11 as the base environment.
``` bash
FROM python:3.11-slim
```
<br> Docker will take a ready-made Python image.

+ Create or use a folder named /app inside the container.
``` bash
WORKDIR /app
```

+  Copy main.py from your laptop into the container’s /app folder.
``` bash
COPY main.py .
```
<br> The dot . means current working folder inside container, which is /app.

+ When the container starts, run this command: `python main.py`

``` bash
CMD ["python", "main.py"]

```

**Step 4: Build Docker image**

Now come back to folder: `docker-python-demo`

``` bash
docker build -t python-demo .
```

| Part             | Meaning                      |
| ---------------- | ---------------------------- |
| `docker build`   | Create Docker image          |
| `-t python-demo` | Name the image `python-demo` |
| `.`              | Use current folder files     |


**Step5: Check images and run it**

``` bash
docker images
docker run python-demo:latest  
```

**Dockerfile does not run directly. First we build an image from Dockerfile, then we run a container from that image.**
