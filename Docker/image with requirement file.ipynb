### What is requirements.txt?

requirements.txt is a file where we write Python package names required by our project.
When Docker builds the image, it reads this file and installs the packages.

**Step 1: create folder**

go to folder in which you want to have dockerfiles stored


``` bash
cd /d E:/data/Documents/Data_Engineering/Module3_Docker

mkdir requirement-demo
cd requirement-demo
```

**step2: Create Python file (main.py) using vs code**


**Step 3: Create requirements.txt**

we require pandas, so this file will tell docker to install pandas

``` bash
pandas==2.2.3
```

**Step 4: Create Dockerfile**

``` bash
FROM python:3.11-slim

WORKDIR /app

COPY requirements.txt .

RUN pip install -r requirements.txt

COPY main.py .

CMD ["python", "main.py"]

```

**Note: CMD runs at container start time, not image build time.**


**Step 4: Build Docker image**

`

``` bash
docker build -t requirement-demo .
```

**Step 5: check whether image is created**

``` bash
docker images
```

**Step6: Run the image**

``` bash
docker run requirement-demo
```
