A Docker volume is a storage area managed by Docker.It is used to store data outside the container’s writable layer.
*Volume keeps data safe even if the container is deleted*

**Step 1: create folder**

go to folder in which you want to have dockerfiles stored


``` bash
cd /d E:/data/Documents/Data_Engineering/Module3_Docker

mkdir docker-volume-demo
cd docker-volume-demo
```

**step2: Create Python file (main.py) using vs code**

``` python
from datetime import datetime
import os

os.makedirs("/data", exist_ok=True)

with open("/data/output.txt", "a") as file:
    file.write(f"Container ran at: {datetime.now()}\n")

print("Data written successfully to /data/output.txt")
```

**Step 3: Create Dockerfile**

``` bash
FROM python:3.11-slim

WORKDIR /app

COPY main.py .

CMD ["python", "main.py"]

```

**Step 4: Build Docker image**

`

``` bash
docker build -t volume-demo-app .
```


**Step 5: Create Docker volume**

Now create a volume:

``` bash
docker volume create my-app-data
```

| Part          | Meaning                  |
| ------------- | ------------------------ |
| `docker`      | Use Docker               |
| `volume`      | Work with Docker volumes |
| `create`      | Create a new volume      |
| `my-app-data` | Name of the volume       |

**check volume**

``` bash
docker volume ls
```

**Volume is created outside the container and managed by Docker.**

**Step-6: Run container with volume**

``` bash
docker run -v my-app-data:/data volume-demo-app

```

| Part                   | Meaning                         |
| ---------------------- | ------------------------------- |
| `my-app-data`          | Docker volume name              |
| `/data`                | Folder inside container         |
| `-v my-app-data:/data` | Attach volume to `/data` folder |
|  `volume-demo-app`      | Image |

**Step7: show output file content**

``` bash
docker run --rm -v my-app-data:/data python:3.11-slim cat /data/output.txt
```

| Part                   | Meaning                                                              |
| ---------------------- | -------------------------------------------------------------------- |
| `docker run`           | Start a temporary container                                          |
| `--rm`                 | Automatically delete this temporary container after command finishes |
| `-v my-app-data:/data` | Attach your volume to `/data` folder                                 |
| `python:3.11-slim`     | Use Python image only to read the file                               |
| `cat /data/output.txt` | Show file content                                                    |

**cat is a Linux command used to show the content of a file in the terminal.**
